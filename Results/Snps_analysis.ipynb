```python
for i,pos in enumerate(snps.ref_pos):
    snp_in_ft = False
    for j,ft in enumerate(sacha.features):
        if ft.type not in {"source","repeat_region"}:
            if pos in ft:
                print(f"snp {i} with pos {pos} in {ft.type} {ft.qualifiers['gene'][0]} with {ft.location} and index {j}")
                snp_in_ft = True
        else:
            continue
    if not snp_in_ft:
        print(f"snp {i} with pos {pos} is not in any feature\n")
    else:
        print("")
```

```python
snp_ft_type = {}
snp_ft_name = {}
snps_wo_ft = 0
for i,pos in enumerate(snps.ref_pos):
    snp_in_ft = False
    for j,ft in enumerate(sacha.features):
        if ft.type not in {"source","repeat_region"}:
            if pos in ft:
                snp_ft_type[i] = ft.type
                snp_ft_name[i] = ft.qualifiers["gene"][0]
                #print(f"snp {i} with pos {pos} in {ft.type} {ft.qualifiers['gene'][0]} with {ft.location} and index {j}")
                snp_in_ft = True
        else:
            continue
    if not snp_in_ft:
        snps_wo_ft += 1 
        #print(f"snp {i} with pos {pos} is not in any feature\n")
    #else:
        #print("")
```

```python
chino = SeqIO.read("../Data/ChinaSI_genbank.gb","gb")
```

```python
from Bio import Align
aligner = Align.PairwiseAligner()
aligner.open_gap_score = -5
aligner.extend_gap_score = -0.5

for ft in chino.features:
    if ft.type == "CDS" and ft.qualifiers["gene"][0] == "rpoC1":
        query = ft.extract(chino).seq

for ft in sacha.features:
    if ft.type == "CDS" and ft.qualifiers["gene"][0] == "rpoC1":
        target = ft.extract(sacha).seq

alns = aligner.align(target,query)
print(f"rpoC1 alignment with {alns.score} score - target {len(target)} - query {len(query)}")
print(alns[0])
```

### __Ricinus__

```python
from Bio import Align
aligner = Align.PairwiseAligner()
aligner.open_gap_score = -5
aligner.extend_gap_score = -0.5

for ft in ricinus.features:
    if ft.type == "CDS" and ft.qualifiers["gene"][0] == "rpoC1":
        query = ft.extract(ricinus).seq.translate(table=11, cds=True)
        rc = ft.extract(ricinus).translate(table=11, cds=True)

for ft in sacha.features:
    if ft.type == "CDS" and ft.qualifiers["gene"][0] == "rpoC1":
        target = ft.extract(sacha).seq.translate(table=11, cds=True)

#print(target)
#print(query)
        
alns = aligner.align(target,query)
print(f"rpoC1 alignment with {alns.score} score - target {len(target)} - query {len(query)}")
print(alns[0])
```

### __Jatropha__

```python
from Bio import Align
aligner = Align.PairwiseAligner()
aligner.open_gap_score = -5
aligner.extend_gap_score = -0.5

for ft in jatropha.features:
    if ft.type == "CDS" and ft.qualifiers["gene"][0] == "rpoC1":
        query = ft.extract(jatropha).seq.translate(table=11, cds=True)
        jc = ft.extract(jatropha).translate(table=11, cds=True)

for ft in sacha.features:
    if ft.type == "CDS" and ft.qualifiers["gene"][0] == "rpoC1":
        target = ft.extract(sacha).seq.translate(table=11, cds=True)
        si = ft.extract(sacha).translate(table=11, cds=True)

#print(target)
#print(query)
        
alns = aligner.align(target,query)
print(f"rpoC1 alignment with {alns.score} score - target {len(target)} - query {len(query)}")
print(alns[0])
```

```python
rpos = [jc,rc,ch,si]
rpos[0].id = rpos[0].desription = rpos[0].name = "jatropha"
rpos[1].id = rpos[1].desription = rpos[1].name = "ricinus"
rpos[2].id = rpos[2].desription = rpos[2].name = "chino"
rpos[3].id = rpos[3].desription = rpos[3].name = "sacha"
SeqIO.write(rpos,"rpoC1.fasta","fasta")
```
